```python
#Pune was represented by ywo Team Names as 'Rising Pune Supergiant' & 'Rising Pune Supergiants' so as a convinience we will changes these with the recent team representing Pune 'Rising Pune Supergiant' in all columns involving this name i.e 'Team' column.
#Similarly 2nd Change is in team name of Delhi ,Earlier the team name for delhi was 'Delhi Daredevils' but later it was changed to 'Delhi Capitals' so we will replace the "delhi Daredevils' with 'Delhi Capitals'.
ipl_df2.Team.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals'},inplace=True)
ipl_df2.Team.unique()
```

```python
# to convet string data type to int to do static calculations
columns =['Matches', 'Batting Innings', 'Not Out',
       'Runs Scored','Batting Average', 'Balls Faced', 'Batting Strike Rate',
       'Wickets Taken', 'Bowling Average','Bowling Economy Rate',
       'Bowling Strike Rate', 'Catches Taken']
ipl_df2[columns] = ipl_df2[columns].astype(float).round().astype(int)
```

```python
 # What is the total number of runs scored by each team in the IPL 2017 tournament?
run_by_team = ipl_df2[ipl_df2['Tournament'] == 'IPL 2017'].groupby('Team')['Runs Scored'].sum()
run_by_team
```

```python
#which team has scored max in IPL 2017 
run_by_team.idxmax()
```

```python
#What is the average batting strike rate for each player across all tournaments?
avg_strike_rate = ipl_df2.groupby('Player')['Batting Strike Rate'].mean()
avg_strike_rate
```

```python
#who has the worst batting strike rate among players and what it is
a = avg_strike_rate.idxmin()
b = avg_strike_rate.min()
'Lowest Batting striker is %s and his strike value is %s'%(a,b)
```

```python
# What is the overall bowling economy rate for each team?
bowling_economy_by_team = ipl_df2.groupby('Team')['Bowling Economy Rate'].mean()

bowling_economy_by_team
```

```python
# What is the correlation between Batting Average and Bowling Average?
correlation_avg = ipl_df2['Batting Average'].corr(ipl_df2['Bowling Average'])

correlation_avg
```

```python
#  Who has the highest batting average in IPL 2019
highest_avg_odi = ipl_df2[ipl_df2['Tournament'] == 'IPL 2019'].nlargest(1, 'Batting Average')

highest_avg_odi
```

```python
# Create a heatmap of the correlation matrix for numerical columns.
correlation_matrix = ipl_df2.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
```

```python
# Create subplots showing the distribution of Runs Scored and Bowling Average.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

sns.histplot(ipl_df2['Runs Scored'], bins=20, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Runs Scored')
axes[0].set_xlabel('Runs Scored')
axes[0].set_ylabel('Frequency')

sns.histplot(ipl_df2['Bowling Average'], bins=20, kde=True, ax=axes[1], color='orange')
axes[1].set_title('Distribution of Bowling Average')
axes[1].set_xlabel('Bowling Average')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
```

```python
#  Create a scatter plot between Wickets Taken and Bowling Economy Rate.
sns.scatterplot(x='Wickets Taken', y='Bowling Economy Rate', data=ipl_df2, color='green')
plt.title('Scatter Plot: Wickets Taken vs Bowling Economy Rate')
plt.xlabel('Wickets Taken')
plt.ylabel('Bowling Economy Rate')
plt.tight_layout()
```

```python
#  How many players have taken more than 5 wickets in a single match?
players_more_than_20_wickets = ipl_df2[ipl_df2['Wickets Taken'] > 20]['Player'].unique()

players_more_than_20_wickets
```

```python
#  Create a bar chart showing the total number of catches taken by each team.

catches_by_team = ipl_df2.groupby('Team')['Catches Taken'].sum()
catches_by_team.plot(kind='bar')
plt.title('Total Catches Taken by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Catches')
plt.show()
```

```python
#  Calculate and display the top 3 players with the highest number of Not Out innings.
top_not_out_players = ipl_df2.nlargest(3, 'Not Out')[['Player', 'Not Out']]

top_not_out_players
```
